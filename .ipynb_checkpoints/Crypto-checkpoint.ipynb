# ***CRYPTOCURRENCY ANALYSIS***

```python
# import dependencies
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
```

```python
# read in crypto csv
df = pd.read_csv('crypto_data.csv')
df = pd.DataFrame(df)
df
```

## clean data for model

```python
# filter for False IsTrading, then drop them. 
df_clean = df[df.IsTrading != False]
df_clean
```

```python
# find nulls
for column in df_clean.columns:
    print(f"Column {column} has {df[column].isnull().sum()} null values")
```

```python
# drop nulls
df_clean = df_clean.dropna()
```

```python
# find any TotalCoinsMined >0
df_clean.loc[df_clean['TotalCoinsMined'] == 0]
```

```python
df_clean.loc[df_clean['IsTrading'] == False]
```

```python
# delete them from data set
df_clean = df_clean[df_clean.TotalCoinsMined != 0]
df_clean
```

```python
# drop CoinName column
df_clean = df_clean.drop(["CoinName"], axis='columns')
df_clean
```

```python
# use get dummies to convert non numerical values
df_clean2 = pd.get_dummies(df_clean, dtype=float)
df_clean2
```

```python
# drop IsTrading Column
df_clean2 = df_clean2.drop(["IsTrading"], axis='columns')
df_clean2
```

## Scale Data for Model

```python
# standard scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(df_clean2)
```

```python
df_clean_scaled = scaler.transform(df_clean2)
df_clean_scaled
```

```python
# initialize PCA model
pca = PCA(n_components=3)
```

```python
# Get three principal components for data
pca_df = pca.fit_transform(df_clean_scaled)
pca_df
```

```python
# transform PCA data to a df
crypto_pca_df = pd.DataFrame(
    data=pca_df, columns=["PC1", "PC2", "PC3"]
)
crypto_pca_df.head()
```

```python
# initialize t-SNE model
tsne = TSNE(learning_rate=35)
```

```python
# Reduce dimensions
tsne_features = tsne.fit_transform(crypto_pca_df)
```

```python
# The dataset has 2 columns
tsne_features.shape
```

```python
# The first column of transformed features
crypto_pca_df['x'] = tsne_features[:,0]

# The second column of transformed features
crypto_pca_df['y'] = tsne_features[:,1]
```

```python
# Visualize the clusters
plt.scatter(crypto_pca_df['x'], crypto_pca_df['y'])
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
# Finding the best value for k
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(crypto_pca_df)
    inertia.append(km.inertia_)

# Creating the Elbow Curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(list(range(11)))
plt.title('Elbow Curve')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# Initializing model with K = 5
model = KMeans(n_clusters=5, random_state=5)
model
```

```python
# Fitting model
model.fit(crypto_pca_df)
```

```python
# Get the predictions
predictions = model.predict(crypto_pca_df)
print(predictions)
```

```python
# Add the predicted class column to the dataframe
crypto_pca_df["class"] = model.labels_
crypto_pca_df.head()
```

```python
# Visualize the clusters
plt.scatter(x=crypto_pca_df['PC1'], y=crypto_pca_df['PC2'], c=crypto_pca_df['class'])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Class')
plt.show()
```
